# Scatter Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading tips dataset from Seaborn
tips = sns.load_dataset("tips")

# Creating scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Setting plot title and axis labels
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

# Showing the plot
plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.regplot(x='total_bill', y='tip', data=tips, color='purple')

plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

# Line Plot

```python
# Creating some sample data
df = pd.DataFrame({
    'Year': ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'],
    'Sales': [1000, 1200, 1500, 1800, 2000, 2200, 2400, 2600, 2800, 3000],
    'Profit': [50, 70, 100, 130, 150, 170, 190, 210, 230, 250]})

# Setting Seaborn style and font scale
sns.set(style='darkgrid', font_scale=1.2)

# Creating a figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plotting the Sales line
sns.lineplot(x='Year', y='Sales', data=df, color='blue', linewidth=1.5, linestyle='-', marker='o', markersize=6, label='Sales')

# Plotting the Profit line
sns.lineplot(x='Year', y='Profit', data=df, color='red', linewidth=1.5, linestyle='--', marker='*', markersize=10, label='Profit')

# Setting axis labels and title
ax.set(xlabel='Year', ylabel='Amount', title='Sales and Profit over Time')

# Adding legend
plt.legend()

# Showing the plot
plt.show()
```

# Bar Plot

```python
# Creating some sample data
df = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [5000, 7000, 4000, 6000],
    'Profit': [1000, 2000, 1500, 2500]})

# Setting Seaborn style and font scale
sns.set(style='darkgrid', font_scale=1.2)

# Creating a figure and axes
fig, ax = plt.subplots(figsize=(7, 5))

# Plotting the Sales bars
sns.barplot(x='Region', y='Sales', data=df, color='black', alpha=0.8, label='Sales')

# Plotting the Profit bars
sns.barplot(x='Region', y='Profit', data=df, color='white', alpha=0.6, label='Profit')

# Setting axis labels and title
ax.set(xlabel='Region', ylabel='Amount', title='Sales and Profit by Region')

# Adding legend
plt.legend()

# Showing the plot
plt.show()
```

# Histogram

```python
# Loading diamonds dataset from Seaborn
diamonds = sns.load_dataset("diamonds")

# Setting Seaborn style and font scale
sns.set(style='whitegrid', font_scale=1.5)

# Creating a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plotting the histogram of diamond prices
sns.histplot(x='price', data=diamonds, color='purple', alpha=0.9, bins=30)

# Setting axis labels and title
ax.set(xlabel='Price', ylabel='Count', title='Distribution of Diamond Prices')

# Showing the plot
plt.show()
```

# Box Plot

```python
# Creating some sample data
np.random.seed(42)
data = pd.DataFrame({'group': np.repeat(['A', 'B', 'C'], 100),
                     'score': np.concatenate([np.random.normal(loc=10, scale=3, size=100),
                                              np.random.normal(loc=20, scale=5, size=100),
                                              np.random.normal(loc=30, scale=2, size=100)])})

# Setting Seaborn style and font scale
sns.set(style='whitegrid', font_scale=1.2)

# Creating a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plotting the box plot of scores by group
sns.boxplot(x='group', y='score', data=data, palette='Set1')

# Setting axis labels and title
ax.set(xlabel='Group', ylabel='Score', title='Box Plot of Scores by Group')

# Showing the plot
plt.show()
```

# Violin Plot

```python
# Loading tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Creating a violin plot
sns.set_style('whitegrid')
sns.violinplot(x='day', y='total_bill', data=tips, color='grey')

# Setting plot labels and title
plt.xlabel('Day of the week')
plt.ylabel('Total Bill')
plt.title('Distribution of Total Bill by Day of the Week')

# Showing the plot
plt.show()
```

# Heatmap

```python
# Loading flights dataset from Seaborn
flights = sns.load_dataset("flights")

# Pivoting the data to create a matrix for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Creating the heatmap using Seaborn
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=.5, ax=ax)

# Adding labels to the plot
ax.set_title("Passenger numbers (1000s) by year and month")
ax.set_xlabel("Year")
ax.set_ylabel("Month")

# Showing the plot
plt.show()
```

# Pair Plot

```python
# Loading the iris dataset
iris = sns.load_dataset("iris")

# Creating a pair plot
sns.pairplot(iris, hue="species", palette="Set1", diag_kind="kde", height=3)

# Setting the title of the plot
plt.suptitle("Pair Plot of Iris Dataset", y=1.05, fontsize=18)

# Showing the plot
plt.show()
```
